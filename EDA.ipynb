```python
#Dataset read without redundant rows using the skip function
df1=pd.read_csv('/content/Absenteeism_at_work.csv',skiprows=25,header=None)
df1
```

```python
# Column headers added to the dataset
absenteeism_column_headers = ['Individual_ID','Reason_for_absence','Month_of_absence','Day_of_the_week',
'Seasons','Transportation_expense','Distance_from_Residence_to_Work','Service_time',
'Age','Work_load_Average/day','Hit_target','Disciplinary_failure','Education',
'Number_of_children','Social_drinker','Social_smoker','Pet','Weight','Height',
'Body_mass_index','Absenteeism_time_in_hours']

df1.columns= absenteeism_column_headers
df1.head()
```

```python
#checking the dataset information
df1.info()
```

```python
#data types conversion - converting other columns to integer
integer_columns = ['Individual_ID', 'Reason_for_absence', 'Month_of_absence', 'Day_of_the_week',
                   'Seasons', 'Transportation_expense', 'Distance_from_Residence_to_Work', 'Service_time',
                   'Age', 'Hit_target', 'Disciplinary_failure', 'Education',
                   'Number_of_children', 'Social_drinker', 'Social_smoker', 'Pet', 'Weight', 'Height',
                   'Body_mass_index', 'Absenteeism_time_in_hours']

df1[integer_columns] = df1[integer_columns].astype(int)

df1['Work_load_Average/day'] = df1['Work_load_Average/day'].astype(float)

df1.info()
```

**Step 4 - Data Visualization**

```python
#Function for defing bar text
def barlabels(bars,c='black'):
    for bar in bars:
        yval = round(bar.get_height(),2)
        plt.text(bar.get_x()+0.25,yval, yval,verticalalignment='bottom',color=c,fontweight='bold')
```

```python
arr=r['Absenteeism_time_in_hours'].count()
arr=np.array(arr)
print(arr)
fig, ax = plt.subplots(figsize=(10,10))
table=[
    'No Reason given','Certain infectious and parasitic diseases',
'Neoplasms',
'blood-forming organs and involving the immune mechanism',
'Endocrine, nutritional and metabolic diseases',
'Mental and behavioural disorders',
'Diseases of the nervous system',
'Diseases of the eye and adnexa',
'Diseases of the ear and mastoid process',
'Diseases of the circulatory system',
'Diseases of the respiratory system',
'Diseases of the digestive system',
'Diseases of the skin and subcutaneous tissue',
'Diseases of the musculoskeletal system and connective tissue',
'Diseases of the genitourinary system',
'Pregnancy, childbirth and the puerperium',
'Certain conditions originating in the perinatal period',
'Congenital malformations, deformations and chromosomal abnormalities',
'Symptoms, signs and abnormal clinical and laboratory findings',
'Injury, poisoning and certain other consequences of external causes',
'Factors influencing health status and contact with health services.',
'patient follow-up',
'medical consultation',
'blood donation',
'laboratory examination',
'unjustified absence',
'physiotherapy',
'dental consultation']
plt.barh(y=np.arange(len(arr)),width=arr,label='No. of instances',color='#00C6C5')
plt.yticks(np.arange(len(arr)),table,rotation=0)
plt.ylabel('Reason of Absence')
plt.xlabel('Count of instances')
plt.title('Reason vs Count',fontweight='bold')
plt.legend()
for i, v in enumerate(arr):
    ax.text(v+2, i, str(v), color='black',fontweight='bold')
plt.show()
```

Addressing the underlying causes of absenteeism related to medical and dental consultations, as well as musculoskeletal disorders, can contribute to improving employee health and well-being, as well as reducing overall absenteeism rates. The company can consider implementing initiatives such as:

```python
plt.figure(figsize=(10,5))
r= df1.groupby('Month_of_absence')
arr=r['Absenteeism_time_in_hours'].count()
arr=np.array(arr)
arr=arr[1:]
arr1=r['Absenteeism_time_in_hours'].mean()
arr1=np.array(arr1)
arr1=arr1[1:]
bars=plt.bar(x=np.arange(len(arr)),height=arr,color='#E6D3E3',label='Count')
bars1=plt.bar(x=np.arange(len(arr)),height=arr1,color='#461C7C',label='Mean')
plt.xlabel('Month of year')
plt.xticks(np.arange(len(arr)),['JAN','FEB','MAR','APR','MAY','JUN','JUL','AUG','SEP','OCT','NOV','DEC'])
plt.title('Variation due to month')
plt.legend()
barlabels(bars)
barlabels(bars1,'red')
plt.show()
```

```python
plt.figure(figsize=(7,7))
r= df1.groupby('Day_of_the_week')
arr=r['Absenteeism_time_in_hours'].count()
arr=np.array(arr)
print(arr)
arr1=r['Absenteeism_time_in_hours'].mean()
arr1=np.array(arr1)
bars=plt.bar(x=np.arange(len(arr)),height=arr,color='grey',label='Count')
bars1=plt.bar(x=np.arange(len(arr)),height=arr1,color='k',label='Mean')
plt.xlabel('Days of week')
plt.xticks(np.arange(len(arr)),['MON','TUE','WED','THUR','FRI'])
plt.title('Days of week variation')
plt.legend()
barlabels(bars)
barlabels(bars1,'white')
plt.show()
```

Mondays typically exhibit higher absenteeism rates compared to other weekdays. This could be attributed to factors such as post-weekend blues, reluctance to return to work after leisure time, or health issues that may have arisen over the weekend.

```python
plt.figure(figsize=(7,7))
r= df1.groupby('Seasons')
arr=r['Absenteeism_time_in_hours'].count()
arr=np.array(arr)
print(arr)
arr1=r['Absenteeism_time_in_hours'].mean()
arr1=np.array(arr1)
bars=plt.bar(x=np.arange(len(arr)),height=arr,color='k',label='Count')
bars1=plt.bar(x=np.arange(len(arr)),height=arr1,color='blue',label='Mean')
plt.xlabel('Season')
plt.xticks(np.arange(len(arr)),['Summer','Autumn','Winter','Spring'])
plt.title('Season variation')
plt.legend()
barlabels(bars)
barlabels(bars1,'white')
plt.show()
```

**Season Variation**

 Spring has the highest number of absenteeisn with 189 employees absent, followed by winter with 177 employees absent, followed by Autumn with 173 absent and summer has the lowest number of employyes with 167 employees ansent
